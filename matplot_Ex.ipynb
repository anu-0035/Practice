```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,5,6,7]
plt.plot(x,y,ls="-.",color="g",lw="1.5",marker="s",mfc="r",mec="b")
plt.xlabel("X- axix")
plt.ylabel("Y- axix")
plt.title("line Plot")
plt.show()
```

```python
import numpy as np
y1=np.array([12,4,5,6,0])
y2=np.array([1,2,3,4,5,6])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,5,6,7]
plt.scatter(x,y,color="r")
plt.xlabel("X- axix")
plt.ylabel("Y- axix")
plt.title("scatter Plot")
plt.show()
```

```python
c=["A","B","C","D","E"]
d=[4,5,9,2,3]
plt.bar(c,d,color="grey",width=0.5)
plt.xlabel("Product")
plt.ylabel("sales (in 1000)")
plt.title("bar graph")
plt.show()
```

```python
c=["A","B","C","D","E"]
d=[4,5,9,2,3]
plt.barh(c,d)
plt.xlabel("Product")
plt.ylabel("sales (in 1000)")
plt.title("bar graph")
plt.show()
```
